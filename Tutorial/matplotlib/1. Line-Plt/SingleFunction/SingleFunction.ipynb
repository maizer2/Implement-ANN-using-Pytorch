```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x - 0), color='black', linestyle='solid')
ax.plot(x, np.sin(x - 1), color='black', linestyle='dashed')
ax.plot(x, np.sin(x - 2), color='black', linestyle='dashdot')
ax.plot(x, np.sin(x - 3), color='black', linestyle='dotted')
```

```python
plt.plot(x, np.sin(x - 0), color='blue', linestyle='-')     # solid
plt.plot(x, np.sin(x - 1), color='green', linestyle='--')   # dashed
plt.plot(x, np.sin(x - 2), color='black', linestyle='-.')   # dashdot
plt.plot(x, np.sin(x - 3), color='red', linestyle=':')      # dotted
plt.plot(x, np.sin(x - 4), color='pink', linestyle='solid')
plt.plot(x, np.sin(x - 5), color='black', linestyle='dashed')
```

```python
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r')
```
