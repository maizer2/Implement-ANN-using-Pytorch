```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x))
plt.title("title")
plt.xlabel("x")
plt.ylabel("y=sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x), '--r')
ax.set_ylabel('y=sin(x)')
ax.set_xlabel('x')
ax.set_title('sin(x)')
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title="y = sin(x)")
```
